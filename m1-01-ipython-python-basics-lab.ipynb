```python
#task 4
events = [
    {"user_id": 101, "event_type": "login", "duration_seconds": 5},
    {"user_id": 102, "event_type": "upload", "duration_seconds": 120},
    {"user_id": 101, "event_type": "update", "duration_seconds": 45},
    {"user_id": 103, "event_type": "login", "duration_seconds": -10}, 
    {"user_id": 104, "event_type": "upload", "duration_seconds": "invalid"}, 
    {"user_id": 102, "event_type": "update", "duration_seconds": 30},
    {"user_id": 105, "event_type": "login", "duration_seconds": 8},
    {"user_id": 101, "event_type": "upload", "duration_seconds": 300},
    {"user_id": 106, "event_type": "login", "duration_seconds": 12},
    {"user_id": 103, "event_type": "update", "duration_seconds": 60},
    {"user_id": 105, "event_type": "upload", "duration_seconds": 150},
    {"user_id": 104, "event_type": "login", "duration_seconds": 4}
]

cleaned_events = []
for event in events:
    duration = event["duration_seconds"]
    
   
    if type(duration)==int or type(duration)==float and duration > 0:
        clean_record = event.copy()
        clean_record["duration_minutes"] = duration / 60
        cleaned_events.append(clean_record)

#Verification
print(f"Original records: {len(events)}")
print(f"Cleaned records: {len(cleaned_events)}")

# Inspecting two cleaned records
cleaned_events[:2]
```

```python
cleaned_events[:2]
```

```python
#Task 5
counts = {}  
totals = {} 
users = set()
for ev in cleaned_events:
    etype = ev["event_type"]

    counts[etype] = counts.get(etype, 0) + 1
    totals[etype] = totals.get(etype, 0) + ev["duration_minutes"]
    
    users.add(ev["user_id"])

csv_lines = ["metric,key,value"] # CSV-nin başlığı (Header)

for etype in counts:
    csv_lines.append(f"count,{etype},{counts[etype]}")
    
    avg = totals[etype] / counts[etype]
    csv_lines.append(f"average_duration_minutes,{etype},{avg:.2f}")


csv_lines.append(f"unique_users,total,{len(users)}")
```
