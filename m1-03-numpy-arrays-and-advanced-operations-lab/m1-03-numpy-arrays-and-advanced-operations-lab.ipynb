```python
#Task1
import numpy as np
rng = np.random.default_rng(seed=0)
readings= rng.normal(50, 20 ,size=(360,4))

readings
```

```python
print(readings[:3])
print(readings.shape)
print(readings.dtype)
```

```python
#Task2
readings_min = readings.min()
readings_max = readings.max()
scaled_readings = (readings - readings_min) / (readings_max - readings_min)

col_means = readings.mean(axis=0) 
centered_readings = readings - col_means

print(f"Orijinal : {readings.shape}")
print(f"Scaled :   {scaled_readings.shape}")
print(f"Centered : {centered_readings.shape}")
print(f"Scaled_Readings : {scaled_readings}")

print(f"\nScaled Min/Max: {scaled_readings.min():.2f} / {scaled_readings.max():.2f}")

print(f"Verified: { scaled_readings.shape == readings.shape and centered_readings.shape == readings.shape }")



```

```python
#task3
mean_val = readings.mean()
std_val = readings.std()
outlier_mask = (readings < 20) | (readings > 80)

outlier_count = np.sum(outlier_mask)
print(f"Outliers: {outlier_count}")

total_elements = readings.size 
total_elements

outlier_percentage = (outlier_count / total_elements) * 100
print(f"Outlier_Percentage : {outlier_percentage}")

cleaned_readings = readings.copy()
cleaned_readings[outlier_mask] = np.nan

nan_count = np.isnan(cleaned_readings).sum()
print(f"NaN count: {nan_count}")
print(f"Eligibility confirmed: {nan_count == outlier_count}")

```

```python
#Task 4
sensor_means = readings.mean(axis=0)
sensor_stds = readings.std(axis=0)

time_means = readings.mean(axis=1)

print("--- Shape Validation ---")
print(f"sensor_means shape: {sensor_means.shape}") 
print(f"sensor_stds shape:  {sensor_stds.shape}")  
print(f"time_means shape:   {time_means.shape}")   


print("\n--- Summary Samples ---")
print(f"Sensor Means: {sensor_means}")
print(f"First 5 Time-step Means: {time_means[:5]}")
```

```python
#Task5
validation_report = {
    "total_readings": readings.size,
    "outlier_count": outlier_count,
    "outlier_percent": outlier_percentage,
    "sensor_means": sensor_means,
    "sensor_stds": sensor_stds
}

size_check = validation_report["total_readings"] == readings.size


print("--- Validation Report ---")
for key, value in validation_report.items():
    if isinstance(value, (float, np.float64)):
        print(f"{key}: {value:.2f}")
    else:
        print(f"{key}: {value}")

print(f"\nValidation (total_readings == readings.size): {size_check}")
```

```python
#Task6
rng = np.random.default_rng(seed=0)

rng = np.random.default_rng(seed=0)
readings= rng.normal(50, 20 ,size=(360,4))

sim = rng.normal(0, 1, size=(1000, 6))

print("--- Simulation Matrix ---")
print(f"Shape: {sim.shape}")
print(f"Dtype: {sim.dtype}")
print("First 3 rows:\n", sim[:3])
```

```python
#Task7

scenario_shift = np.array([-0.3, -0.1, 0.0, 0.1, 0.2, 0.4])

adjusted_sim = sim + scenario_shift

print("\n--- Broadcasting Validation ---")
print(f"Adjusted shape matches original: {adjusted_sim.shape == sim.shape}")


print("Original column means:", sim.mean(axis=0))
print("Adjusted column means:", adjusted_sim.mean(axis=0))
```

```python
#Task8

scenario_means = adjusted_sim.mean(axis=0)

ranked_indices = np.argsort(scenario_means)

partitioned_means = np.partition(scenario_means, -2)

top_two_partition = partitioned_means[-2:]


full_sort_top_two = np.sort(scenario_means)[-2:]

match_check = top_two_partition==full_sort_top_two


print("\n--- Ranking and Partitioning ---")
print(f"Scenario Rankings (indices): {ranked_indices}")
print(f"Top 2 Means (via partition): {top_two_partition}")
print(f"Matches full sort: {match_check}")
```

```python
#Task9

rng_2 = np.random.default_rng(seed=0)
sim_2 = rng_2.normal(0, 1, size=(1000, 6))


rng_3 = np.random.default_rng(seed=123)
sim_3 = rng_3.normal(0, 1, size=(1000, 6))


same_check = np.array_equal(sim, sim_2)
diff_check = not np.array_equal(sim, sim_3)

print("\n--- Reproducibility ---")
print(f"Same seed matches exactly: {same_check}")
print(f"Different seed produces different results: {diff_check}")
```

```python
#Task10
report_2 = {
    "shape": adjusted_sim.shape,
    "scenario_means": scenario_means,
    "top_two_indices": ranked_indices[-2:].tolist(),
    "top_two_values": top_two_partition.tolist()
}

print("\n--- Validation Report ---")
for key, value in report_2.items():
    print(f"{key}: {value}")


means_len_check = len(report_2["scenario_means"]) == 6
indices_len_check = len(report_2["top_two_indices"]) == 2

print(f"\nValidation - Scenario means length is 6: {means_len_check}")
print(f"Validation - Top two indices length is 2: {indices_len_check}")
```
